```python
#Linear Regression in pytorch

X_train = np.arange(10, dtype = 'float32').reshape((10, 1))
y_train = np.array([1.0, 1.3, 3.1, 2.0, 5.0, 6.3, 6.6, 7.4, 8.0, 9.0], dtype='float32')
```

```python
plt.plot(X_train, y_train, 'o', markersize = 10)
```

```python
from torch.utils.data import TensorDataset

X_trainNorm = (X_train - np.mean(X_train))/np.std(X_train)
X_trainNorm = torch.from_numpy(X_trainNorm)

y_train = torch.from_numpy(y_train).float()

trainDS = TensorDataset(X_trainNorm, y_train)

batchSize = 1

trainDL = DataLoader(trainDS, batchSize, shuffle=True)
```

```python
torch.manual_seed(1)
weight = torch.randn(1)
weight.requires_grad_()
bias = torch.zeros(1, requires_grad=True)
#Requires grad tells the computer that a gradient needs to calculates 
```

```python
def model(xb):
    return xb @ weight + bias
```

```python
#MSE
def lossFN(input, target):
    return (input - target).pow(2).mean()
```

```python
for x, y in trainDL:
    print(x, y)
```

```python
pred = model()
```

```python
#Using stochastic gradient descent to find the weights

learningRate = 0.001
numEpochs = 200
logEpochs = 10

for epoch in range(numEpochs):
    for xBatch, yBatch in trainDL:
        pred = model(xBatch) # xBatch*weight + bias. So the predicted ting
        loss = lossFN(pred, yBatch.long()) #Calculating the loss. .long() -> equivalent to putting int64
        loss.backward() #Backward computes the gradient
    
    with torch.no_grad(): #With no_grad -> changes the grad_required to False
        weight -= weight.grad * learningRate
        bias -= bias.grad * learningRate
        weight.grad.zero_()
        bias.grad.zero_()
        
    if epoch%logEpochs == 0:
        print(epoch, loss.item())
```

```python
#Using torch.nn and torch.optim to minimize the loss function MSE using stochastic gradient descent

import torch.nn as nn

lossFN = nn.MSELoss(reduction='mean')
inputSize = 1
outputSize = 1
model = nn.Linear(inputSize, outputSize)

optimizer = torch.optim.SGD(model.parameters(), lr = 0.0001)
```

```python
for epoch in range(numEpochs):
    for xBatch, yBatch in trainDL:
        pred = model(xBatch)[:, 0] #Generate Prediction
        loss = lossFN(pred, yBatch) #The loss function
        loss.backward() #Compute gradients
        optimizer.step() #Update the parameters using gradients
        optimizer.zero_grad()
        
    if epoch % logEpochs == 0:
        print(epoch, loss.item())
        
```

```python
#Using pytorch for classification

from sklearn.datasets import load_iris

from sklearn.model_selection import train_test_split

iris = load_iris()

X = iris['data']
y = iris['target']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=1)
```

```python
X_trainNorm = (X_train - np.mean(X_train))/np.std(X_train)
X_trainNorm = torch.from_numpy(X_trainNorm).float()

y_train = torch.from_numpy(y_train)

trainDS = TensorDataset(X_trainNorm, y_train)

torch.manual_seed(1)

batchSize = 2

trainDL = DataLoader(trainDS, batch_size=2, shuffle = True)
```

```python
# Building 2 layers
# The first layer is Linear layer, f(w*x + b)
# The first layer recieves 4 dimensions, outputs 16 dimensions
# The second layer takes in the 16 dimensions, and outputs 3 neurons, as we have three class labels

class Model(nn.Module):
    
    def __init__(self, inputSize, hiddenSize, outputSize):
        super().__init__()
        
        self.layer1 = nn.Linear(inputSize, hiddenSize)
        self.layer2 = nn.Linear(hiddenSize, outputSize)
        
    def forward(self, x):
        x = self.layer1(x)
        x = nn.Sigmoid()(x)
        x = self.layer2(x)
        
        return x
```

```python
inputSize = X_trainNorm.shape[1]
hiddenSize = 16
outputSize = 3

model = Model(inputSize, hiddenSize, outputSize)
```

```python
#Loss function as cross-entropy loss, and the optimizer is Adam

learningRate = 0.001
loss_fn = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(), lr = 0.001)
```

```python
#Training the Model

numEpochs = 100
loss_hist = [0]*numEpochs
accuracy_hist = [0]*numEpochs

for epoch in range(numEpochs):
    for xBatch, yBatch in trainDL:
        pred = model(xBatch)
        loss = loss_fn(pred, yBatch)
        loss.backward()
        optimizer.step()
        optimizer.zero_grad()
        
        
```

```python
import matplotlib.pyplot as plt

plt.plot(loss_hist)
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.show()

plt.plot(accuracy_hist)
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.show()
```
